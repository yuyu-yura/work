```python
import folium
```

```python
# 위도
latitude = 37.394946
# 경도
longitude = 127.111104
```

```python
m = folium.Map(location=[latitude, longitude],
               zoom_start=17, 
               width=750, 
               height=500
              )
m
```

```python
folium.Marker([latitude, longitude],
              popup="판교역",
              tooltip="판교역 입구").add_to(m)
m
```

```python
folium.Marker([latitude, longitude],
              popup="판교역",
              tooltip="판교역 입구", 
              icon=folium.Icon('red', icon='star'),
             ).add_to(m)
m
```

```python
import pandas as pd
```

```python
df = pd.read_csv('광주아파트1.csv')
df.head(3)
```

```python
pd.Series(df.columns)
```

```python
sub_df = df.loc[df['소재지도로명주소'].isin(['광산구', '동구', '북구'])]
sub_df.head(3)
```

```python
folium.CircleMarker([35.1597648, 126.851544],
                    color='tomato',
                    radius = 50, 
                    tooltip='광주시청 상권').add_to(m)
m
```

```python
from folium.plugins import MarkerCluster


m = folium.Map(
    location=[latitude, longitude],
    zoom_start=15
)

coords = sub_df[['위도', '경도']]


marker_cluster = MarkerCluster().add_to(m)

for lat, long in zip(coords['위도'], coords['경도']):
    folium.Marker([lat, long], icon = folium.Icon(color="green")).add_to(marker_cluster)
m
```
