```python
import folium
```

```python
latitude = 33.453361
longitude = 126.571111
```

```python
import pandas as pd
```

```python
df=pd.read_csv("금호2.csv")
df.head(3)
```

```python
pd.Series(df.columns)
```

```python
df[['위도','경도']]
```

```python
from folium.plugins import MarkerCluster
m = folium.Map(
    location=[latitude, longitude],
    zoom_start=50
)

coords = df[['위도', '경도']]
marker_cluster = MarkerCluster().add_to(m)

for lat, long in zip(coords['위도'], coords['경도']):
  folium.Marker([lat, long], icon=folium.Icon('orange', icon='flower')).add_to(m)
m.save('금호2.html')
m
```
