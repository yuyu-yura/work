```python
import folium
```

```python
latitude = 33.457361
longitude = 126.564806
```

```python
import pandas as pd
```

```python
df=pd.read_csv("금호1.csv")
df.head(3)
```

```python
pd.Series(df.columns)
```

```python
df[['위도','경도']]
```

```python
from folium.plugins import MarkerCluster
m = folium.Map(
    location=[latitude, longitude],
    zoom_start=15
)

coords = df[['위도', '경도']]

marker_cluster = MarkerCluster().add_to(m)

for lat, long in zip(coords['위도'], coords['경도']):
  folium.Marker([lat, long], icon=folium.Icon('red', icon='home')).add_to(marker_cluster)
m.save('금호.html')
m
```
